```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

from vector import Vector, print_Vector
from line import Line
from line_vortex import biot_savart
```

Make the line vortex extremely long

```python
L = 1e5
p0 = Vector(0, 0, -L/2)
p1 = Vector(0, 0,  L/2)
line = Line(p0, p1)
```

## Example along x-axis

```python
n = 11
x = np.linspace(0, 1, n)
y = np.zeros_like(x)
z = np.zeros_like(x)
us = np.zeros(n)
vs = np.zeros(n)
ws = np.zeros(n)
for i in range(n):
    xv = Vector(x[i], y[i], z[i])
    uvv = biot_savart(xv, line)
    us[i], vs[i], ws[i] = uvv.ravel()
```

```python
plt.figure()
plt.quiver(x, y, us, vs)
plt.xlim(0, 1.2)
plt.ylim(-0.1, 0.1)
plt.tick_params(direction='in')
```

## Example in the x-y plane

```python
n = 21
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y, indexing='ij')
us = np.zeros((n, n))
vs = np.zeros((n, n))
ws = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        xv = Vector(X[i, j], Y[i, j], 0.0)
        uvv = biot_savart(xv, line)
        us[i, j], vs[i, j], ws[i,j] = uvv.ravel()
```

```python
plt.figure()
plt.quiver(X, Y, us, vs)
```
