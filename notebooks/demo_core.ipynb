```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

import OpenAEM
```

```python
origin = np.array([1, 0, 0])
eddy = OpenAEM.Attached_Eddy(method='Pi')
eddy_mirror = eddy.mirror()
axs = plt.figure(figsize=(10, 10)).add_subplot(projection='3d')
eddy.plot(axs)
eddy_mirror.plot(axs, color='yellow')

axs.set_xlabel('x')
axs.set_ylabel('y')
axs.set_zlabel('z')
axs.set_xlim(0, 3)
axs.tick_params(direction='in')
```

```python
origin = np.array([1, 0, 0])
eddy = OpenAEM.Attached_Eddy.from_Lambda(origin=origin)
eddy_mirror = eddy.mirror()
axs = plt.figure(figsize=(10, 10)).add_subplot(projection='3d')
eddy.plot(axs)
eddy_mirror.plot(axs, color='yellow')

axs.set_xlabel('x')
axs.set_ylabel('y')
axs.set_zlabel('z')
axs.set_xlim(0, 3)
axs.tick_params(direction='in')
```
