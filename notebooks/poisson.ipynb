```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
patch = OpenAEM.Wall_Patch(xmax=2.0, zmax=2.0)
lams = [64, 16, 4, 1]
sizes = [4, 8, 16, 32]
colors = ['black', 'red', 'blue', 'green']
for i in range(len(lams)):
    pts = patch.place_eddies(lam=lams[i], seed=111)
    ax.plot(pts[0, :], pts[1, :], linestyle='none', marker='s', markerfacecolor='none', markeredgecolor=colors[i], markersize=sizes[i], label=f'Hierarchey {i+1}')
    
ax.legend()
```
